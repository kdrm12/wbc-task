```python
import torch as th
th.has_mps
device = "mps" if th.has_mps else "cpu"
print(f'Using device: {device}')
import torchvision.transforms as T
import requests
from PIL import Image, ImageDraw, ImageFont
from ultralytics import YOLO
import numpy as np
import supervision as sv
```

```python
model=YOLO('yolov9.pt')
model.train(data='D:\experimen\dataset\DeTR.v9i.yolov8\data.yaml',epochs=100)
```

```python
dataset=sv.DetectionDataset.from_yolo(images_directory_path=r'D:\experimen\dataset\DeTR.v9i.yolov8\train\images',annotations_directory_path=r'D:\experimen\dataset\DeTR.v9i.yolov8\train\labels',data_yaml_path=r'D:\experimen\dataset\DeTR.v9i.yolov8\data.yaml')
modelcm=YOLO(r'D:\experimen\ADR\runs\detect\train3\weights\best.pt')

def callback(image: np.ndarray) -> sv.Detections:
    result=modelcm(image)[0]
    return sv.Detections.from_ultralytics(result)

confmatr=sv.ConfusionMatrix.benchmark(
    dataset=dataset,
    callback=callback
)

confmatr.plot()
```
